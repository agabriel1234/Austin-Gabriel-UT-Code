### TASK 2
- Task 2.1: Write a Function that Finds Eigenvalues and Eigenvectors
 - Use numpy.linalg.eig to compute the eigenvalues and eigenvectors for each matrix.
- Task 2.2: Plot Eigenvalue Distribution
 - Plot the eigenvalues of each matrix using a scatter plot or line plot.
 - This will help visualize how the eigenvalues are distributed and give insight into whether the matrix is positive definite.


```python
def find_eigvalue_eigvector(matrices):
    eigenvalues_list = []
    eigenvectors_list = []
    for matrix in matrices:
        eigenvalues, eigenvectors = np.linalg.eig(matrix)
        eigenvalues_list.append(eigenvalues)
        eigenvectors_list.append(eigenvectors)
    return eigenvalues_list, eigenvectors_list

# Calculate eigenvalues and eigenvectors for the 10 matrices
eigenvalues_list, eigenvectors_list = find_eigvalue_eigvector(matrices)

# Display the eigenvalues of each matrix
for i, eigenvalues in enumerate(eigenvalues_list):
    print(f"Eigenvalues of Matrix A[{i+1}]:\n{eigenvalues}\n")

# Task 2.2: Plot the eigenvalue distribution for each matrix
def plot_eigenvalue_distribution(eigenvalues_list):
    plt.figure(figsize=(10, 6))
    for i, eigenvalues in enumerate(eigenvalues_list):
        plt.scatter(range(1, len(eigenvalues) + 1), eigenvalues, label=f'Matrix {i+1}')
    plt.xlabel('Eigenvalue Index')
    plt.ylabel('Eigenvalue')
    plt.title('Eigenvalue Distribution of the 10 Random Matrices')
    plt.legend()
    plt.show()

# Plot the eigenvalue distribution
plot_eigenvalue_distribution(eigenvalues_list)

#This plot will exclude imaginary eigenvalues
```

### TASK 3
- Task 3.1: Write a function that Checks if the Matrix is Diagonally Dominant
    - A matrix is diagonally dominant if for each row, the magnitude of the diagonal element is greater than or equal to the sum of the magnitudes of the other elements in that row.
- Task 3.2: More than likely, none of the random matrices will be diagnially dominant, modify your 10th matrix and *make* it diagonally dominant by changing the diagonal elements so they are greater than the sum of the rest of the elements in the row

```python
#Task 3.1
def is_diagonally_dominant(matrix):
    for i in range(len(matrix)):
        row_sum = np.sum(np.abs(matrix[i, :])) - np.abs(matrix[i, i])
        if np.abs(matrix[i, i]) < row_sum:
            return False
    return True

diagonal_dominance_results = [is_diagonally_dominant(matrix) for matrix in matrices]

for i, is_dominant in enumerate(diagonal_dominance_results):
    print(f"Matrix A[{i+1}] is diagonally dominant: {is_dominant}")

#Task 3.2
def make_diagonally_dominant(matrix):
    new_matrix = matrix.copy()
    for i in range(len(matrix)):
        row_sum = np.sum(np.abs(matrix[i, :])) - np.abs(matrix[i, i])
        new_matrix[i, i] = row_sum + 1  # Make the diagonal element larger than the sum of the other elements
    return new_matrix

# Modify the 10th matrix
matrices[9] = make_diagonally_dominant(matrices[9])

# Check if the 10th matrix is now diagonally dominant
is_dominant_after_modification = is_diagonally_dominant(matrices[9])
print(f"Matrix A[10] after modification is diagonally dominant: {is_dominant_after_modification}")
```

```python
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy() 
    errors = []
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        
        error = np.linalg.norm(x_new - x, ord=np.inf)
        errors.append(error)  # Append the error at each iteration
        
        # Check for convergence
        if error < tol:
            return x_new, k, errors
        x = x_new
    return x, max_iterations, errors


def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    errors = []  # To track error at each iteration
    
    for k in range(max_iterations):
        x_new = x.copy()
 
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))  # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))  # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        
        error = np.linalg.norm(x_new - x, ord=np.inf)
        errors.append(error)  # Append the error at each iteration

        if error < tol:
            return x_new, k, errors
 
        x = x_new
 
    return x, max_iterations, errors


initial_tol = 1e-3
max_iterations = 100

def Gauss_Jacobi(initial_tol, max_iterations, matrices, vectors, A, b, x0):
    gsolutions = []
    jsolutions = []
    gerrors = []
    gitlist = []
    gtollist = []  # Separate list for tolerances for Gauss-Seidel (not truly needed for this instance - legacy code of mine)
    jerrors = []
    itlist = []
    tollist = []  # Separate list for tolerances for Jacobi
    for idx, (A,b) in enumerate(zip(matrices, vectors)):
        print(f"Solving for Matrix {idx + 1}")

        x0 = np.zeros_like(b)

        # Jacobi method
        tol = initial_tol
        solution, iterations, errors = jacobi(A, b, x0, tol, max_iterations)
        jerrors.append(errors)
        itlist.append(iterations)

        jsolution = solution
        print(f"Solution Jacobi x =  \n{solution}")
        print(f"Iterations Jacobi: \n{iterations}")
        
        jsolutions.append(jsolution)
        
        # Gauss-Seidel method
        solution, iterations, errors = gauss_seidel(A, b, x0, tol, max_iterations)
        gerrors.append(errors)
        gitlist.append(iterations)

        gsolution = solution    
        print(f"Solution Gauss x =  \n{solution}")
        print(f"Iterations Gauss: \n{iterations}")
        
        gsolutions.append(gsolution)
    
    # Return after the loop has completed for all matrices
    return gsolutions, jsolutions, jerrors, itlist, gerrors, gitlist, tollist, gtollist

x0 = np.zeros_like(b)

# Now call the function to solve all matrices
gsolutions, jsolutions, jerrors, itlist, gerrors, gitlist, tollist, gtollist = Gauss_Jacobi(initial_tol, max_iterations, matrices, vectors, A, b, x0)
```
