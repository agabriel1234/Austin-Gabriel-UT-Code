```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters

k = 1.0 # Spring constant
h_values = [.01, .02, .03, .05, .1, .2, .3, .5] # Different step sizes to test stability
t_start = 0 # Start time
t_end = 20 # End time
y_0 = 1.0 # Initial displacement
v_0 = 0.0 # Initial velocity

# Function to apply Runge Kutta 4's Method for given step size



def runge_kutta_4_2nd_order(h, k, t_start, t_end, y_0, v_0):
    t_values = np.arange(t_start, t_end + h, h)
    y_values = np.zeros(len(t_values))
    v_values = np.zeros(len(t_values))
    
    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0
    
    for i in range(1, len(t_values)):
        t_n = t_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]
        
        # Calculate RK4 increments for y and v
        k1_y = h * v_n
        k1_v = h * (-k * y_n)
        
        k2_y = h * (v_n + k1_v / 2)
        k2_v = h * (-k * (y_n + k1_y / 2))
        
        k3_y = h * (v_n + k2_v / 2)
        k3_v = h * (-k * (y_n + k2_y / 2))
        
        k4_y = h * (v_n + k3_v)
        k4_v = h * (-k * (y_n + k3_y))
        
        # Update y and v using RK4 formula
        y_values[i] = y_n + (k1_y + 2 * k2_y + 2 * k3_y + k4_y) / 6
        v_values[i] = v_n + (k1_v + 2 * k2_v + 2 * k3_v + k4_v) / 6
        
    return t_values, y_values



# Plot results for different step sizes

plt.figure(figsize=(12, 6))
for h in h_values:
    t, y = runge_kutta_4_2nd_order(h, k, t_start, t_end, y_0, v_0)
    plt.plot(t, y, label=f'Runge Kutta 4\'s Method (h={h})')
plt.xlabel('Time (t)')
plt.ylabel('Displacement (y)')
plt.title('Stability Analysis of Runge Kutta 4\'s Method for a 2nd Order ODE k = 1')
plt.legend()
plt.grid(True)
plt.show()

#This graph remains relatively stable - step size seems to have little effect as well, with displacement oscillating from -1 to 1 over the span of 0 to 20 seconds
#Therefore it is not becoming unstable any time soon
```
